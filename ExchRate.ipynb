```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import dates as mpl_dates
from IPython.display import display, clear_output


plt.rcParams['figure.figsize'] = 15, 10
plt.rcParams['font.size'] = 20
```

```python
Data = pd.read_csv("Interbank FX Rates Historical.csv",index_col=0, parse_dates=True)
```

```python
Data=Data[(Data['Buying']!=0) & Data['Selling']!=0]
```

```python
Selling_Rate= pd.to_numeric(Data["Selling"])
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1) 

date_format = mpl_dates.DateFormatter('%d-%m-%Y')
plt.gcf().autofmt_xdate()
plt.gca().xaxis.set_major_formatter(date_format)

plt.xlabel('time')
plt.ylabel('GHS/CNY')
plt.title('Cedi devaluation against the Chinese RMB since 2013')
    
for i in range(len(Selling_Rate)):
    x=Selling_Rate[-(i+1):]
    
    ax.plot(1/x,'b')
    
    display(fig)
    
    clear_output(wait = True)
```
